### Data Science Bootcamp


```python
df_ds = df_top15[['query','url','order','rank','key_count','word_count']][df_top15['query']=='Data Science Bootcamp']
```

```python
df_ds['url_sub']=df_ds['url']
#df_ds['url_sub']=df_ds['url_sub'].str.slice(23, )[df_ds['url_sub']!='https://weclouddata.com/']
df_ds['url_sub']
```

```python
df_ds['url_sub']=df_ds['url_sub'].str.slice(23, )
df_ds['url_sub']
```

```python
df_ds 
```

```python
from matplotlib.pyplot import figure
```

```python
GA = nx.from_pandas_edgelist(df_ds, source="query", target="url_sub")
figure(figsize=(20, 10))
nx.draw(GA, with_labels=True, node_color='skyblue', node_size=1000)
```
